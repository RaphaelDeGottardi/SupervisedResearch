```python
import pandas as pd
from pysyncon import Dataprep, Synth
```

```python
us_data = pd.read_csv('merged_trends_data_US.csv', parse_dates=True, index_col='date')
brex_data = pd.read_csv('google_trends_data_bref.csv', parse_dates=True, index_col='date')
ca_data = pd.read_csv('merged_trends_data_CA.csv', parse_dates=True, index_col='date')
au_data = pd.read_csv('merged_trends_data_AU.csv', parse_dates=True, index_col='date')
nz_data = pd.read_csv('merged_trends_data_NZ.csv', parse_dates=True, index_col='date')
```

```python
keywords = [
    "Astrology",
    "Horoscope",
    "Fortune telling",
    "Psychic",
    "Tarot reading",
    "Numerology",
    "Palm reading",
    "Clairvoyant",
    "Crystal ball",
    "Magic spells",
    "Occult",
    "Spiritual healing",
    "Witchcraft",
    "Mediumship",
    "Superstition"
]
```

```python
control_data = [us_data, ca_data, au_data, nz_data]
control_combined = pd.concat(control_data, axis=1, keys=[f'country_{i}' for i in range(len(control_data))])
```

```python
def prepare_combined_data(keyword, brex_data, control_combined):
    treated = brex_data[keyword].rename("value")
    treated = treated.reset_index()
    treated['unit'] = 'UK'

    control = control_combined.xs(keyword, level=1, axis=1)
    control = control.reset_index().melt(id_vars='date', var_name='unit', value_name='value')

    combined_data = pd.concat([treated, control], ignore_index=True)
    
    # Ensure date column is of datetime type
    combined_data['date'] = pd.to_datetime(combined_data['date'])
    combined_data['time'] = combined_data['date'].astype('int64')
    
    return combined_data
```
